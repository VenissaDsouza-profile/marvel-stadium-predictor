```python
# ============================================
# PROJECT 1: Marvel Stadium Attendance Predictor
# Notebook 1: Data Collection
# ============================================

import requests
import pandas as pd
import os

# Create data folder if it doesn't exist
os.makedirs("data", exist_ok=True)

print("✅ All libraries loaded successfully!")
```

```python
API_KEY = "e7oUjUDP8tp5bit2AC6G8Y9e0GF0dKze"

url = "https://app.ticketmaster.com/discovery/v2/events.json"
params = {
    "apikey": API_KEY,
    "venueId": "KovZpaooxe",  # ✅ Correct Marvel Stadium ID
    "countryCode": "AU",
    "size": 200
}

response = requests.get(url, params=params)
data = response.json()

events = data["_embedded"]["events"]
df_events = pd.DataFrame([{
    "name": e.get("name"),
    "date": e["dates"]["start"].get("localDate"),
    "genre": e["classifications"][0]["segment"]["name"] if e.get("classifications") else None,
    "url": e.get("url")  
} for e in events])

df_events.to_csv("data/marvel_events.csv", index=False)
print(f"✅ Saved {len(df_events)} events!")
df_events.head()
```

```python
url = "https://archive-api.open-meteo.com/v1/archive"
params = {
    "latitude": -37.8136,
    "longitude": 144.9631,
    "start_date": "2022-01-01",
    "end_date": "2024-12-31",
    "daily": "temperature_2m_max,precipitation_sum,windspeed_10m_max",
    "timezone": "Australia/Melbourne"
}

response = requests.get(url, params=params)
data = response.json()

df_weather = pd.DataFrame(data["daily"])
df_weather.to_csv("data/melbourne_weather.csv", index=False)
print(f"✅ Saved {len(df_weather)} days of weather data!")
df_weather.head()
```

```python
url = "https://afltables.com/afl/stats/biglists/bg3.txt"
df_afl = pd.read_csv(url, sep="\t", on_bad_lines="skip")
df_afl.to_csv("data/afl_games.csv", index=False)
print(f"✅ Saved {len(df_afl)} AFL records!")
df_afl.head()
```

```python
dfs = []
for year in ["2022", "2023", "2024"]:
    url = f"https://date.nager.at/api/v3/PublicHolidays/{year}/AU"
    response = requests.get(url)
    dfs.append(pd.DataFrame(response.json()))

df_holidays = pd.concat(dfs, ignore_index=True)
df_holidays = df_holidays[["date", "name"]]
df_holidays.to_csv("data/public_holidays.csv", index=False)
print(f"✅ Saved {len(df_holidays)} public holidays!")
df_holidays.head()
```
